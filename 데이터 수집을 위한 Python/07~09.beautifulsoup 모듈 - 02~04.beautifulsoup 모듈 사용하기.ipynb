* id, class 속성으로 tag 찾기
 - 타이틀
 - 작성자, 작성일

```python
url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=104&oid=421&aid=0005410696'
resp = requests.get(url)

# resp.text
soup = BeautifulSoup(resp.text)
title = soup.find('h3', id="articleTitle" ) #<h3 id="articleTitle" class="tts_head">마크롱 "미국 돌아왔나"…바이든 "물론"</h3>
title.get_text()
```

```python
url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=104&oid=421&aid=0005410696'
resp = requests.get(url)

# resp.text
soup = BeautifulSoup(resp.text)
title = soup.find('h3')
title.get_text()
```

```python
url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=104&oid=421&aid=0005410696'
resp = requests.get(url)

soup = BeautifulSoup(resp.text)
container = soup.find('div', id='articleBodyContents') 
container.get_text()
```

```python
url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=104&oid=421&aid=0005410696'
resp = requests.get(url)

soup = BeautifulSoup(resp.text)
container = soup.find('div', id='articleBodyContents') 
contents = ''
for p in container.find_all('span'):
    contents += p.get_text()
    
contents
```

* CSS를 이용하여 tag 찾기
 - select, select_one함수 사용 
 - css selector 사용법
   - 태그명 찾기 tag 
   - 자손 태그 찾기 - 자손 관계 (tag tag)
   - 자식 태그 찾기 - 다이렉트 자식 관계 (tag > tag)
   - 아이디 찾기 #id
   - 클래스 찾기 .class
   - 속성값 찾기 [name='test']
     - 속성값 prefix 찾기 [name ^='test']
     - 속성값 suffix 찾기 [name $='test']
     - 속성값 substring 찾기 [name *='test]
   - n번째 자식 tag 찾기 :nth-child(n)
